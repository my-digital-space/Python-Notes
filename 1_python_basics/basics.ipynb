# Syntax of Python program

## Introduction

```python
print("Hello World 1!")
print("Hello World 2!")
print("Hello World 3!")
```

In Python, `print()` is a built-in function used to display output to the console (terminal).

What `print()` does
- It takes one or more values
- Converts them to strings (if needed)
- Sends them to standard output (screen)

👉 By default it <span style="color: orange;">adds a newline</span> at the end.

## 🐍 Python syntax rules

1️⃣ Indentation

Python uses indentation (spaces) to define blocks of code instead of `{}` like Java or C.

✅ Example

```python
x = 10

if x > 5:
    print("Inside if")   # indented → inside block
    print("Still inside if")

print("Outside if")      # not indented → outside block
```

🔑 Key points
- Indented lines belong to the block
- Same indentation = same block level
- Wrong indentation → IndentationError

👉 Think of indentation as:

> “Am I inside this condition/loop/function or not?”

2️⃣ Whitespace (formatting matters)
Python is sensitive to whitespace, especially indentation.

✅ Best practice
- Use <span style="color: orange;">4 spaces</span> (PEP 8 standard)
- Do NOT mix tabs and spaces
- ❌ Bad example

```python
if True:
 print("Wrong")   # inconsistent indentation
```

✅ Good example

```python
if True:
    print("Correct")
```

3️⃣ Statements (lines of code)

A statement is basically one instruction.

✅ Normal usage

```python
a = 10
b = 20
print(a + b)
```

⚠️ Possible but <span style="color: orange;">not recommended</span>

```python
a = 10; b = 20; print(a + b)
```

👉 Why avoid this?

Harder to read
Goes against Python’s philosophy: “Readability counts”

4️⃣ Comments

Comments are ignored by Python — they are for humans.

✅ Single-line comment

```python
# This is a comment
print("Hello")
```

✅ Multi-line comments (using multiple #)

Python doesn’t have a special syntax for multi-line comments, so we use multiple `#`.

✅ Example

```python
# This is a multi-line comment
# explaining multiple things
# about the code
print("Hello")
```

# Python Variables

## Introduction

A variable in Python is a <span style="color: orange;">name</span> (identifier) given to a value stored in memory.

👉 In simple terms:

> A variable is like a <span style="color: orange;">label for data</span> so we can use and change it in a program.

🧠 Simple example

```python
name = "John"
age = 34
```

- `name` → variable name
- `"John"` → value stored
- `age` → variable name
- `34` → value stored

👉 Python stores these values in memory and we use the variable names to access them.

⚙️ How variables are created

In Python, a variable is created when we assign a value using `=`

```python
x = 10
```

- `x` → variable
- `=` → assignment operator
- `10` → value

👉 No need to declare type separately.

## 🔥 Dynamic typing

Dynamic typing means:
- We do NOT need to declare the data type of a variable.  
- Python automatically detects the type based on the value assigned.

🧠 Simple example

```python
x = 10
print(type(x))   # int

x = "Hello"
print(type(x))   # str
```

👉 The same variable `x`:
- First stores an integer
- Then stores a string

Python allows this because it is dynamically typed.

🔁 Changing type anytime

```python
value = 100        # int
value = 3.14       # float
value = "Python"   # string
```

👉 This is completely valid in Python.

⚠️ Important disadvantage

Errors can occur at runtime

```python
x = "10"
y = 5
print(x + y)   # ❌ TypeError
```

## Rules of defining a variable in Python

1️⃣ Variable name must start with a letter or underscore (_)

It cannot start with a number

```python
valid_name = 10
_name = 20

2name = 30  # ❌ Invalid
```

2️⃣ Variable name can contain letters, digits, and underscores

Only these characters are allowed: `a-z`, `A-Z`, `0-9`, `_`

```python
name1 = "John"
user_age = 25

user-age = 30  # ❌ Invalid (hyphen not allowed)
```

3️⃣ Variable names are <span style="color: orange;">case-sensitive</span>

```python
age = 25
Age = 30

print(age)  # 25
print(Age)  # 30
```

4️⃣ Variable name cannot be a Python keyword

Reserved words like `if`, `for`, `while`, `class`, etc. cannot be used

```python
for = 10  # ❌ Invalid
class = "Test"  # ❌ Invalid
```

5️⃣ No special characters allowed (except underscore)

```python
user@name = "John"  # ❌ Invalid
price$ = 100        # ❌ Invalid
```

## Best practices for naming variables in Python

1️⃣ Use descriptive and meaningful names

Variable names should clearly indicate what data they hold

```python
user_age = 25      # ✅ clear
x = 25             # ❌ unclear
```

2️⃣ Use lowercase letters

Variable names are typically written in lowercase

```python
total = 100        # ✅
Total = 100        # ❌ (not recommended)
```

3️⃣ Use underscores to separate words (snake_case)

Improves readability for multi-word variable names

```python
first_name = "John"
total_amount = 500
```

4️⃣ Keep names concise but readable

Avoid overly long or unnecessarily short names

```python
count = 10             # ✅ good balance
number_of_students_in_classroom = 10  # ❌ too long
c = 10                 # ❌ too short
```

5️⃣ Avoid using single-letter names (except in simple loops)

```python
for i in range(5):   # ✅ acceptable
    print(i)

user_count = 5       # ✅ better for general use
```

6️⃣ Avoid using names that shadow built-in functions

```python
list = [1, 2, 3]   # ❌ shadows built-in 'list'
sum = 10           # ❌ shadows built-in 'sum'
```

7️⃣ Use consistent naming style throughout the code

Stick to one convention (preferably snake_case)

```python
user_name = "Alice"
user_age = 25
```

8️⃣ Use meaningful prefixes/suffixes when needed

```python
is_active = True      # boolean variable
user_list = []        # list variable
total_count = 10
```

## Python Data Types (Built-in)

### 1️⃣ Numeric types

Python provides three built-in numeric data types:

1️⃣ `int`

Represents <span style="color: orange;">whole numbers</span> (positive, negative, or zero)

```python
a = 10
b = -5
c = 0
```

2️⃣ `float`

3️⃣ `complex`

`a` = real part, `b` = imaginary part

Key points
- Python automatically determines the type (dynamic typing)
- We can check the type using `type()`

Python provides the following built-in sequence types:

1️⃣ `str`

- Represents a <span style="color: orange;">sequence of characters</span> (text)

```python
name = "Alice"
```

What does "cannot be changed after creation" mean?

It means once a string is created, we cannot modify its existing characters.

👉 Instead, we have to create a new string:

```python
text = "hello"
text = "H" + text[1:]   # ✅ creates a new string
print(text)  # "Hello"
```

2️⃣ `list`

- Represents an <span style="color: orange;">ordered collection</span> of items

- <span style="color: orange;">Mutable</span> (can be changed)

```python
numbers = [1, 2, 3, 4]
```

3️⃣ `tuple`

- Represents an <span style="color: orange;">ordered collection</span> of items
- <span style="color: orange;">Immutable</span>

```python
coordinates = (10, 20)
```

4️⃣ `range`

- Represents a <span style="color: orange;">sequence of numbers</span>
- Commonly used in loops

```python
r = range(1, 5)   # 1, 2, 3, 4
```

Common properties of sequence types
- Ordered (maintain insertion order)
- Indexed (access using index)
- Iterable (can loop through)

```python
items = [10, 20, 30]

print(items[0])   # 10

for item in items:
    print(item)
```

Indexing and slicing

```python
text = "Python"

print(text[0])     # 'P'
print(text[-1])    # 'n'
print(text[0:3])   # 'Pyt'
```

### 3️⃣ Mapping type

Python provides one built-in mapping type:

1️⃣ `dict`

Stores data in key-value pairs

```python
student = {
    "name": "John",
    "age": 20,
    "marks": 85
}
```

Key characteristics
- Keys are unique
- Keys are immutable types (e.g., `str`, `int`, `tuple`)
- Values can be of any data type
- Dictionaries are mutable (can be changed after creation)

Accessing values

```python
student = {
    "name": "John",
    "age": 20
}

print(student["name"])   # John
```

Adding / updating values

```python
student = {
    "name": "John"
}

student["age"] = 20          # add
student["name"] = "Alice"    # update
```

Removing elements

```python
student = {
    "name": "John",
    "age": 20
}

del student["age"]
```

### 4️⃣ Set types

Python provides two built-in set types:

1️⃣ `set`

Represents an <span style="color: orange;">unordered</span> collection of <span style="color: orange;">unique</span> elements

```python
s = {1, 2, 3, 3, 4}
print(s)   # {1, 2, 3, 4}
```

2️⃣ `frozenset`

Represents an <span style="color: orange;">immutable</span> version of a set

- Unordered (no indexing)

```python
s = {1, 2, 3}
s.add(4)
s.remove(2)
```

Key characteristics of frozenset

Immutable (cannot modify after creation)

```python
fs = frozenset([1, 2, 3])

# fs.add(4) ❌ Error
```

Type checking

```python
print(type({1, 2}))              # <class 'set'>
print(type(frozenset([1, 2])))  # <class 'frozenset'>
```

### 6️⃣ Binary types

Python provides three built-in binary data types:

1️⃣ `bytes`
- Represents an immutable sequence of bytes
- Values range from 0 to 255

```python
b = b"hello"
print(b)          # b'hello'
print(b[0])       # 104 (ASCII value of 'h')
```

2️⃣ `bytearray`
- Represents a mutable sequence of bytes

```python
ba = bytearray(b"hello")
ba[0] = 72        # changing 'h' → 'H'
print(ba)         # bytearray(b'Hello')
```

3️⃣ `memoryview`
- Provides a view of a binary object without copying data

```python
data = bytearray(b"abc")
mv = memoryview(data)

print(mv[0])      # 97

mv[0] = 120       # modify through memoryview
print(data)       # bytearray(b'xbc')
```

Key points
- `bytes` → immutable
- `bytearray` → mutable
- `memoryview` → efficient access without copying data

Type checking

```python
print(type(b"abc"))          # <class 'bytes'>
print(type(bytearray(b"abc")))  # <class 'bytearray'>
print(type(memoryview(b"abc"))) # <class 'memoryview'>
```

## Python `type()` Method

The `type()` function is a built-in Python function with two distinct uses:

1️⃣ Checking the Type of an Object

Syntax: `type(object)`

Returns the type/class of the given object.

```python
print(type(42))          # <class 'int'>
print(type(3.14))        # <class 'float'>
print(type("hello"))     # <class 'str'>
print(type([1, 2, 3]))   # <class 'list'>
print(type({"a": 1}))    # <class 'dict'>
print(type(True))        # <class 'bool'>
```

Checking type in conditions:

```python
x = 42
if type(x) is int:
    print("x is an integer")   # ✅ x is an integer
```

2️⃣ Dynamically Creating a New Class

Syntax: `type(name, bases, dict)`

| Parameter | Description |
|-----------|-------------|
| `name`| Name of the new class (string) |
| `bases`| Tuple of base/parent classes |
| `dict`| Dictionary of attributes and methods |

```python
# Dynamically create a class named "Dog"
Dog = type("Dog", (object,), {"sound": "Woof", "legs": 4})

d = Dog()
print(d.sound)   # Woof
print(d.legs)    # 4
```

This is equivalent to:

```python
class Dog(object):
    sound = "Woof"
    legs = 4
```

## Typecasting in Python

What is typecasting?

Typecasting means converting one data type into another data type

```python
x = "10"
y = int(x)   # string → int
```

Inbuilt typecasting methods

1️⃣ int()
2️⃣ float()
3️⃣ str()
4️⃣ bool()
5️⃣ list()
6️⃣ tuple()
7️⃣ set()
8️⃣ dict()

```python
a = int("10")
b = float(5)
c = str(100)
d = list("abc")
```

Same type to same type conversion

If we convert a value to the same type, nothing changes

```python
x = 10
y = int(x)

print(y)        # 10
print(type(y))  # int
```

Invalid conversion

If conversion is not possible, Python throws an error

```python
x = "abc"
y = int(x)   # ❌ invalid
```

Handling errors using try-except

```python
try:
    x = "abc"
    y = int(x)
    print(y)

except ValueError:
    print("Conversion failed: invalid input")
```

# Operators in Python

## 1️⃣ Arithmetic operators

Arithmetic operators are used to perform basic mathematical operations on values.

```python
# Arithmetic Operators Example

a = 10
b = 3

# Addition
print(a + b)   # 13

# Subtraction
print(a - b)   # 7

# Multiplication
print(a * b)   # 30

# Division (always returns float)
print(a / b)   # 3.333...

# Modulus (remainder)
print(a % b)   # 1

# Exponentiation (power)
print(a ** b)  # 1000

# Floor Division (removes decimal part)
print(a // b)  # 3
```

## 2️⃣ Assignment operators

Assignment operators are used to assign values to variables, and in many cases, to update the variable based on its current value.

Assignment operators update a variable by combining assignment (`=`) with another operation.

```python
# 1️⃣ = (Simple assignment)
# Assigns a value directly to a variable
x = 10   # x becomes 10

# 2️⃣ += (Add and assign)
# Adds a value to the variable and stores the result back
x = 10
x += 5   # x = x + 5 → 15

# 3️⃣ -= (Subtract and assign)
# Subtracts a value from the variable
x = 10
x -= 3   # x = x - 3 → 7

# 4️⃣ *= (Multiply and assign)
# Multiplies the variable by a value
x = 10
x *= 2   # x = x * 2 → 20

# 5️⃣ /= (Divide and assign)
# Divides the variable by a value (result is always float)
x = 10
x /= 4   # x = x / 4 → 2.5

# 6️⃣ %= (Modulus and assign)
# Stores the remainder after division
x = 10
x %= 3   # x = x % 3 → 1

# 7️⃣ **= (Power and assign)
# Raises the variable to a power
x = 2
x **= 3   # x = x ** 3 → 8

# 8️⃣ //= (Floor division and assign)
# Divides and rounds down to nearest integer
x = 10
x //= 3   # x = x // 3 → 3

# 9️⃣ &= (Bitwise AND and assign)
# Performs AND operation on each bit
x = 10    # 1010
x &= 3    # 0011
# Result: 0010 → 2

# 🔟 |= (Bitwise OR and assign)
# Performs OR operation on each bit
x = 2     # 0010
x |= 4    # 0100
# Result: 0110 → 6

# 1️⃣1️⃣ ^= (Bitwise XOR and assign)
# Performs XOR (different bits → 1)
x = 6     # 0110
x ^= 1    # 0001
# Result: 0111 → 7

# 1️⃣2️⃣ >>= (Right shift and assign)
# Shifts bits to the right (divides by 2 each shift)
x = 8     # 1000
x >>= 1
# Result: 0100 → 4

# 1️⃣3️⃣ <<= (Left shift and assign)
# Shifts bits to the left (multiplies by 2 each shift)
x = 3     # 0011
x <<= 2
# Result: 1100 → 12
```

## 3️⃣ Comparison operators

Comparison operators are used to compare two values and return a boolean result (`True` or `False`).

```python
# Comparison Operators Example

a = 10
b = 20

print(a == b)   # Equal to → checks if a is equal to b → False
print(a != b)   # Not equal to → checks if a is not equal to b → True
print(a > b)    # Greater than → checks if a is greater than b → False
print(a < b)    # Less than → checks if a is less than b → True
print(a >= b)   # Greater than or equal to → checks if a is greater than or equal to b → False
print(a <= b)   # Less than or equal to → checks if a is less than or equal to b → True
```

## 4️⃣ Logical operators

Logical operators are used to combine multiple conditions and return a boolean result (`True` or `False`).

Logical operators must be written in <span style="color: orange;">lowercase</span>.

```python
# Logical Operators Example

a = 10
b = 20
c = 10

print(a < b and a == c)
# AND → True only if both conditions are True → (True and True) → True

# print(a < b AND a == c)  # ❌ AND is not valid in Python
#   - Python is case-sensitive
#   - Logical operators must be written in lowercase

print(a < b and a > c)  
# AND → False if any one condition is False → (True and False) → False

print(a > b or a == c)  
# OR → True if at least one condition is True → (False or True) → True

print(a > b or a != c)  
# OR → False only if both conditions are False → (False or False) → False

print(not(a < b))  
# NOT → reverses the result → not(True) → False

print(not(a > b))  
# NOT → reverses the result → not(False) → True
```

## 5️⃣ Bitwise operators

Bitwise operators perform operations on numbers at the binary (bit) level.

```python
# Bitwise Operators Example

a = 5   # Binary: 0101
b = 3   # Binary: 0011

print(a & b)   # AND → 0101 & 0011 = 0001 → 1
print(a | b)   # OR → 0101 | 0011 = 0111 → 7
print(a ^ b)   # XOR → 0101 ^ 0011 = 0110 → 6
print(~a)      # NOT → ~0101 = -(5+1) = -6
print(a << 1)  # Left Shift → 0101 << 1 = 1010 → 10
print(a >> 1)  # Right Shift → 0101 >> 1 = 0010 → 2
```

## 6️⃣ Membership operators

Membership operators are used to check whether a value exists within a sequence (like `list`, `tuple`, `string`, etc.).

```python
# Membership Operators Example

numbers = [1, 2, 3, 4, 5]
text = "hello"

print(3 in numbers)        # in → checks if 3 is present in the list → True
print(10 in numbers)       # in → checks if 10 is present in the list → False

print(3 not in numbers)    # not in → checks if 3 is NOT present → False
print(10 not in numbers)  # not in → checks if 10 is NOT present → True

print('h' in text)         # in → checks if 'h' is present in the string → True
print('z' not in text)     # not in → checks if 'z' is NOT present → True
```

## 7️⃣ Identity operators

Identity operators are used to compare whether two variables refer to the same object in memory.

```python
# Identity Operators Example

a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a is b)       # is → checks if a and b refer to the same object → False
print(a is c)       # is → checks if a and c refer to the same object → True

print(a is not b)   # is not → checks if a and b are NOT the same object → True
print(a is not c)   # is not → checks if a and c are NOT the same object → False
```

# Control flow and loops in Python

## 1️⃣ Conditional statements

Conditional statements allow us to execute different blocks of code based on conditions.

- `if` → checks a condition
- `elif` → checks another condition if previous ones fail
- `else` → runs when none of the above conditions are true

```python
age = 20

# if condition
if age < 13:
    print("Child")

# elif condition (checked only if above is False)
elif age >= 13 and age < 18:
    print("Teenager")

# elif condition
elif age >= 18 and age < 60:
    print("Adult")

# else condition (runs if all above conditions are False)
else:
    print("Senior citizen")
```

## 2️⃣ Looping statements

Looping statements allow us to execute a block of code repeatedly.

- `for` → used to iterate over a sequence (like list, string, range)
- `while` → runs as long as a condition is `True`

```python
# ----------- FOR LOOP -----------
# Iterating over a list
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    print("For loop value:", num)


# Iterating using range()
for i in range(3):
    print("Range loop value:", i)


# ----------- WHILE LOOP -----------
# Loop runs until condition becomes False
count = 0

while count < 3:
    print("While loop value:", count)
    count += 1  # increment to avoid infinite loop
```

⚠️ We must avoid infinite loops, especially in `while` loops, by ensuring the condition eventually becomes `False`.

```python
# ----------- INFINITE LOOP EXAMPLE (DO NOT RUN) -----------
# This loop will run forever because the condition never becomes False

# while True:
#     print("This will run infinitely")  # ❌ No exit condition
```

## 3️⃣ Loop control statements

Loop control statements allow us to alter the normal flow of loops.

- `break` → exits the loop completely
- `continue` → skips the current iteration and moves to the next
- `pass` → does nothing (placeholder statement)

```python
for i in range(1, 6):

    # break example
    if i == 4:
        print("Breaking the loop at", i)
        break   # exits the loop completely

    # continue example
    if i == 2:
        print("Skipping", i)
        continue   # skips rest of this iteration

    # pass example
    if i == 3:
        pass   # does nothing, placeholder for future logic

    print("Current value:", i)
```
