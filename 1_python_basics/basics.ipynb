# Syntax of Python program

## Introduction

```python
print("Hello World 1!")
print("Hello World 2!")
print("Hello World 3!")
```

In Python, `print()` is a built-in function used to display output to the console (terminal).

What `print()` does
- It takes one or more values
- Converts them to strings (if needed)
- Sends them to standard output (screen)

👉 By default it <span style="color: orange;">adds a newline</span> at the end.


---

## 🐍 Python syntax rules

1️⃣ Indentation

Python uses indentation (spaces) to define blocks of code instead of `{}` like Java or C.

✅ Example

```python
x = 10

if x > 5:
    print("Inside if")   # indented → inside block
    print("Still inside if")

print("Outside if")      # not indented → outside block
```

🔑 Key points
- Indented lines belong to the block
- Same indentation = same block level
- Wrong indentation → IndentationError

👉 Think of indentation as:

> “Am I inside this condition/loop/function or not?”

2️⃣ Whitespace (formatting matters)
Python is sensitive to whitespace, especially indentation.

✅ Best practice
- Use <span style="color: orange;">4 spaces</span> (PEP 8 standard)
- Do NOT mix tabs and spaces
- ❌ Bad example

```python
if True:
 print("Wrong")   # inconsistent indentation
```

✅ Good example

```python
if True:
    print("Correct")
```

3️⃣ Statements (lines of code)

A statement is basically one instruction.

✅ Normal usage

```python
a = 10
b = 20
print(a + b)
```

⚠️ Possible but <span style="color: orange;">not recommended</span>

```python
a = 10; b = 20; print(a + b)
```

👉 Why avoid this?

Harder to read
Goes against Python’s philosophy: “Readability counts”

4️⃣ Comments

Comments are ignored by Python — they are for humans.

✅ Single-line comment

```python
# This is a comment
print("Hello")
```

✅ Multi-line comments (using multiple #)

Python doesn’t have a special syntax for multi-line comments, so we use multiple `#`.

✅ Example

```python
# This is a multi-line comment
# explaining multiple things
# about the code
print("Hello")
```

# Python Variables

## Introduction

A variable in Python is a <span style="color: orange;">name</span> (identifier) given to a value stored in memory.

👉 In simple terms:

> A variable is like a <span style="color: orange;">label for data</span> so we can use and change it in a program.

🧠 Simple example

```python
name = "John"
age = 34
```

- `name` → variable name
- `"John"` → value stored
- `age` → variable name
- `34` → value stored

👉 Python stores these values in memory and we use the variable names to access them.

⚙️ How variables are created

In Python, a variable is created when we assign a value using `=`

```python
x = 10
```

- `x` → variable
- `=` → assignment operator
- `10` → value

👉 No need to declare type separately.

## 🔥 Dynamic typing

Dynamic typing means:
- We do NOT need to declare the data type of a variable.  
- Python automatically detects the type based on the value assigned.

🧠 Simple example

```python
x = 10
print(type(x))   # int

x = "Hello"
print(type(x))   # str
```

👉 The same variable `x`:
- First stores an integer
- Then stores a string

Python allows this because it is dynamically typed.

🔁 Changing type anytime

```python
value = 100        # int
value = 3.14       # float
value = "Python"   # string
```

👉 This is completely valid in Python.

⚠️ Important disadvantage

Errors can occur at runtime

```python
x = "10"
y = 5
print(x + y)   # ❌ TypeError
```

## Rules of defining a variable in Python

1️⃣ Variable name must start with a letter or underscore (_)

It cannot start with a number

```python
valid_name = 10
_name = 20

2name = 30  # ❌ Invalid
```

2️⃣ Variable name can contain letters, digits, and underscores

Only these characters are allowed: `a-z`, `A-Z`, `0-9`, `_`

```python
name1 = "John"
user_age = 25

user-age = 30  # ❌ Invalid (hyphen not allowed)
```

3️⃣ Variable names are <span style="color: orange;">case-sensitive</span>

```python
age = 25
Age = 30

print(age)  # 25
print(Age)  # 30
```

4️⃣ Variable name cannot be a Python keyword

Reserved words like `if`, `for`, `while`, `class`, etc. cannot be used

```python
for = 10  # ❌ Invalid
class = "Test"  # ❌ Invalid
```

5️⃣ No special characters allowed (except underscore)

```python
user@name = "John"  # ❌ Invalid
price$ = 100        # ❌ Invalid
```

## Best practices for naming variables in Python

1️⃣ Use descriptive and meaningful names

Variable names should clearly indicate what data they hold

```python
user_age = 25      # ✅ clear
x = 25             # ❌ unclear
```

2️⃣ Use lowercase letters

Variable names are typically written in lowercase

```python
total = 100        # ✅
Total = 100        # ❌ (not recommended)
```

3️⃣ Use underscores to separate words (snake_case)

Improves readability for multi-word variable names

```python
first_name = "John"
total_amount = 500
```

4️⃣ Keep names concise but readable

Avoid overly long or unnecessarily short names

```python
count = 10             # ✅ good balance
number_of_students_in_classroom = 10  # ❌ too long
c = 10                 # ❌ too short
```

5️⃣ Avoid using single-letter names (except in simple loops)

```python
for i in range(5):   # ✅ acceptable
    print(i)

user_count = 5       # ✅ better for general use
```

6️⃣ Avoid using names that shadow built-in functions

```python
list = [1, 2, 3]   # ❌ shadows built-in 'list'
sum = 10           # ❌ shadows built-in 'sum'
```

7️⃣ Use consistent naming style throughout the code

Stick to one convention (preferably snake_case)

```python
user_name = "Alice"
user_age = 25
```

8️⃣ Use meaningful prefixes/suffixes when needed

```python
is_active = True      # boolean variable
user_list = []        # list variable
total_count = 10
```

# Python data types

## Python Data Types (Built-in)

### 1️⃣ Numeric types

Python provides three built-in numeric data types:

1️⃣ `int`

Represents <span style="color: orange;">whole numbers</span> (positive, negative, or zero)

```python
a = 10
b = -5
c = 0
```

2️⃣ `float`

Represents <span style="color: orange;">decimal</span> (floating-point) numbers

```python
x = 10.5
y = -3.14
z = 0.0
```

3️⃣ `complex`

Represents complex numbers in the form: `a + bj`

`a` = real part, `b` = imaginary part

```python
c1 = 2 + 3j
c2 = -1 + 4j
```

Key points
- Python automatically determines the type (dynamic typing)
- We can check the type using `type()`

```python
print(type(10))       # <class 'int'>
print(type(10.5))     # <class 'float'>
print(type(2 + 3j))   # <class 'complex'>
```

### 2️⃣ Sequence types

Python provides the following built-in sequence types:

1️⃣ `str`

- Represents a <span style="color: orange;">sequence of characters</span> (text)

- <span style="color: orange;">Immutable</span> (cannot be changed after creation)

```python
name = "Alice"
```

What does "cannot be changed after creation" mean?

It means once a string is created, we cannot modify its existing characters.

```python
text = "hello"

text[0] = "H"   # ❌ Error (not allowed)
```

👉 Instead, we have to create a new string:

```python
text = "hello"
text = "H" + text[1:]   # ✅ creates a new string
print(text)  # "Hello"
```

2️⃣ `list`

- Represents an <span style="color: orange;">ordered collection</span> of items

- <span style="color: orange;">Mutable</span> (can be changed)

```python
numbers = [1, 2, 3, 4]
```

3️⃣ `tuple`

- Represents an <span style="color: orange;">ordered collection</span> of items
- <span style="color: orange;">Immutable</span>

```python
coordinates = (10, 20)
```

4️⃣ `range`

- Represents a <span style="color: orange;">sequence of numbers</span>
- Commonly used in loops

```python
r = range(1, 5)   # 1, 2, 3, 4
```

Common properties of sequence types
- Ordered (maintain insertion order)
- Indexed (access using index)
- Iterable (can loop through)

```python
items = [10, 20, 30]

print(items[0])   # 10

for item in items:
    print(item)
```

Indexing and slicing

```python
text = "Python"

print(text[0])     # 'P'
print(text[-1])    # 'n'
print(text[0:3])   # 'Pyt'
```

### 3️⃣ Mapping type

Python provides one built-in mapping type:

1️⃣ `dict`

Stores data in key-value pairs

```python
student = {
    "name": "John",
    "age": 20,
    "marks": 85
}
```

Key characteristics
- Keys are unique
- Keys are immutable types (e.g., `str`, `int`, `tuple`)
- Values can be of any data type
- Dictionaries are mutable (can be changed after creation)

Accessing values

```python
student = {
    "name": "John",
    "age": 20
}

print(student["name"])   # John
```

Adding / updating values

```python
student = {
    "name": "John"
}

student["age"] = 20          # add
student["name"] = "Alice"    # update
```

Removing elements

```python
student = {
    "name": "John",
    "age": 20
}

del student["age"]
```

Common methods

```python
student = {
    "name": "John",
    "age": 20
}

student.keys()     # dict_keys(['name', 'age'])
student.values()   # dict_values(['John', 20])
student.items()    # dict_items([('name', 'John'), ('age', 20)])
```

### 4️⃣ Set types

Python provides two built-in set types:

1️⃣ `set`

Represents an <span style="color: orange;">unordered</span> collection of <span style="color: orange;">unique</span> elements

```python
s = {1, 2, 3, 3, 4}
print(s)   # {1, 2, 3, 4}
```

2️⃣ `frozenset`

Represents an <span style="color: orange;">immutable</span> version of a set

```python
fs = frozenset([1, 2, 3, 3])
print(fs)   # frozenset({1, 2, 3})
```

Key characteristics of set
- Unordered (no indexing)
- No duplicate values
- Mutable (can add/remove elements)

```python
s = {1, 2, 3}
s.add(4)
s.remove(2)
```

Common set operations

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)  # Union → {1, 2, 3, 4, 5}
print(a & b)  # Intersection → {3}
print(a - b)  # Difference → {1, 2}
```

Key characteristics of frozenset

Immutable (cannot modify after creation)

```python
fs = frozenset([1, 2, 3])

# fs.add(4) ❌ Error
```

Type checking

```python
print(type({1, 2}))              # <class 'set'>
print(type(frozenset([1, 2])))  # <class 'frozenset'>
```

### 5️⃣ Boolean type

1️⃣ `bool`

- Represents truth values
- Only two possible values: `True` and `False`

```python
a = True
b = False
```

Boolean values from expressions

```python
x = 10
y = 5

print(x > y)   # True
print(x < y)   # False
```

Using `bool()` for conversion

```python
print(bool(1))        # True
print(bool(0))        # False
print(bool("Hello"))  # True
print(bool(""))       # False
```

Key points
- `True` and `False` are case-sensitive
- Internally, `True` behaves like `1` and `False` like `0`

```python
print(True + 1)   # 2
print(False + 5)  # 5
```

Common usage

```python
is_logged_in = True
is_admin = False
```

### 6️⃣ Binary types

Python provides three built-in binary data types:

1️⃣ `bytes`
- Represents an immutable sequence of bytes
- Values range from 0 to 255

```python
b = b"hello"
print(b)          # b'hello'
print(b[0])       # 104 (ASCII value of 'h')
```

2️⃣ `bytearray`
- Represents a mutable sequence of bytes

```python
ba = bytearray(b"hello")
ba[0] = 72        # changing 'h' → 'H'
print(ba)         # bytearray(b'Hello')
```

3️⃣ `memoryview`
- Provides a view of a binary object without copying data

```python
data = bytearray(b"abc")
mv = memoryview(data)

print(mv[0])      # 97

mv[0] = 120       # modify through memoryview
print(data)       # bytearray(b'xbc')
```

Key points
- `bytes` → immutable
- `bytearray` → mutable
- `memoryview` → efficient access without copying data

Type checking

```python
print(type(b"abc"))          # <class 'bytes'>
print(type(bytearray(b"abc")))  # <class 'bytearray'>
print(type(memoryview(b"abc"))) # <class 'memoryview'>
```

### 7️⃣ None type

1️⃣ `NoneType`

Represents the <span style="color: orange;">absence</span> of a value or <span style="color: orange;">no</span> value

```python
x = None
```

Key points
- `None` is a special constant in Python
- Its type is `NoneType`

```python
print(type(None))   # <class 'NoneType'>
```

It is often used to initialize variables or indicate “nothing”

```python
result = None
```

Functions that do not return anything implicitly return `None`

```python
def test():
    pass

print(test())   # None
```

Comparison with `None` should be done using `is`

```python
x = None

if x is None:
    print("No value")
```
