# Decorators in Python

## Introduction

1️⃣ The core idea

A decorator is simply:

>A function that takes another function and returns a new function with extra behavior

So instead of modifying an existing function directly, we <span style="color: orange;">wrap it</span>.

2️⃣ Is it a nested function?

Yes — internally, decorators almost always use nested functions (closures).

Why?

Because:
- we need to define a new function inside it
- that inner function can access the original function (`func`)
- and add behavior before/after calling it

Example

```python
def add_logging(func):
    def wrapper():
        print("Starting order process...")
        func()
        print("Order process completed.")
    return wrapper

def place_order():
    print("Order placed successfully")

enhanced_function = add_logging(place_order)
enhanced_function()
```

Here:
- `add_logging` → decorator
- `wrapper` → nested function
- `func` → original function

Note that we did not do any changes to the orginal function `order_food`.

👉 What happens:
- we pass `place_order` into decorator
- decorator returns `wrapper`
- we call the new function



3️⃣ What the wrapper actually becomes

Conceptually, this happens:

```python
def enhanced_function():
    print("Starting order process...")
    print("Order placed successfully")
    print("Order process completed.")
```

4️⃣ What problem does it solve?

Decorators are commonly used when we want to:

✔️ Do something BEFORE a function
- logging
- authentication
- validation

✔️ Do something AFTER a function
- cleanup
- logging result
- sending notifications

✔️ Do BOTH

5️⃣ The shortcut syntax: `@decorator`

Instead of writing:

```python
enhanced_function = add_logging(place_order)
```

We write:

```python
@add_logging
def place_order():
    print("Order placed successfully")
```

👉 Python internally converts this to:

```python
place_order = add_logging(place_order)
```

```python
def add_logging(func): # decorator name → custom (can be any name)
    def wrapper():     # inner function → custom (can be any name)
        print("Starting order process...")
        func()         # parameter → custom (can be any name)
        print("Order process completed.")
    return wrapper

@add_logging           # @ → Python syntax for applying decorator
def place_order():
    print("Order placed successfully")

place_order()
```

Real-world use cases
- Logging
- Authentication
- Authorization
- Caching
- Timing performance
- Input validation

## Decorators with single argument

Decorators with arguments add one more level of nesting.

1️⃣ Basic example (single parameter)

Repeat order confirmation N times

```python
def repeat_order(times):
    def decorator(func):
        def wrapper(arg):
            for _ in range(times):
                func(arg)
        return wrapper
    return decorator

@repeat_order(2)
def confirm_order(order_id):
    print(f"Order {order_id} confirmed")

confirm_order("ORD101")
```

2️⃣ Why extra nesting is needed

When we write:

```python
@repeat(3)
def greet():
    ...
```

Python does:

```python
greet = repeat(3)(greet)
```

So:
- `repeat(3)` → must return a decorator
- that decorator → takes the function
- and returns the wrapper

👉 That’s why we need 3 layers:

<pre>
outer function (takes decorator args)
    ↓
decorator function (takes function)
    ↓
wrapper function (runs actual logic)
</pre>

## Decorators with multile arguments

1️⃣ Example with multiple parameters

Scenario: Logging with custom prefix and level

```python
def log_message(prefix, level):
    def decorator(func):
        def wrapper(*args, **kwargs):
            print(f"[{level}] {prefix} - Start")
            result = func(*args, **kwargs)
            print(f"[{level}] {prefix} - End")
            return result
        return wrapper
    return decorator

@log_message("OrderService", "INFO")
def place_order(item):
    print(f"Placing order for {item}")

place_order("Pizza")
```

### What are `*args` and `**kwargs`?

1️⃣ What are *args and **kwargs?

They are not keywords ❌

They are just conventions (names) ✔️

What actually matters is:
- `*` → collects positional arguments
- `**` → collects keyword arguments

2️⃣ Why do we need them in decorators?

Because decorators should work with ANY function, like:

```python
def f1():
    pass

def f2(x):
    pass

def f3(x, y, z=10):
    pass
```

If we don’t use `*args`, `**kwargs`, our decorator will break for some functions where there will be arguments but our decorators would miss them.

# `*args` and `*kwargs` in Python

1️⃣ What are `*args` and `**kwargs`?

They are not keywords ❌

They are just conventions (names) ✔️

What actually matters is:
- `*` → collects positional arguments
- `**` → collects keyword arguments

2️⃣ What does `*args` mean?

```python
def demo(*args):
    print(args)
    sum = 0
    for num in args:
        sum += num
    print(f"Total sum = {sum}")

demo(1, 2, 3)
demo(1, 2, 3, 4, 5) # So, now our function will still work irrespective of the number of arguments being passed to it
```

👉 So: `*args` collects all <span style="color: orange;">positional arguments</span> into a <span style="color: orange;">tuple</span>

3️⃣ What does `**kwargs` mean?

```python
def demo(**kwargs):
    print(kwargs)
    for item in kwargs.keys():
        print(f"Item= {item}:{kwargs[item]}")

demo(name="Pizza", price=200)
```

👉 So: `**kwargs` collects all <span style="color: orange;">keyword arguments</span> into a <span style="color: orange;">dictionary</span>

4️⃣ Can we rename them?

YES ✅

```python
def demo(*numbers, **details):
    print(numbers)
    print(details)
```

This works perfectly.

👉 So:
- `args` and `kwargs` are just naming conventions (best practice)
- not mandatory names

5️⃣ Why the `*` and `**`?

These are Python syntax operators.
- `*` (single star)
    - collects positional arguments
    - packs them into a tuple
- `**` (double star)
    - collects keyword arguments
    - packs them into a dictionary

6️⃣ Example using both `*args` and `**kwargs`

```python
def process_order(*items, **details):
    print("Items ordered:")
    for item in items:
        print(f"- {item}")
    
    print("\nOrder details:")
    for key, value in details.items():
        print(f"{key}: {value}")

process_order(
    "Pizza", "Burger", "Pasta",
    customer="Rahul",
    table=5,
    payment="UPI"
)
```

What is happening internally

1️⃣ Positional arguments → `*items`

("Pizza", "Burger", "Pasta") # Stored as a tuple

2️⃣ Keyword arguments → `**details`
<pre>
{
    "customer": "Rahul",
    "table": 5,
    "payment": "UPI"
}
</pre> # Stored as a dictionary

Important rules

✔ Order matters

```python
def process_order(*items, **details):
```

`*args` comes before `**kwargs`

✔ Function call structure

```python
process_order("Pizza", "Burger", key="value")
```

positional first, keyword later
