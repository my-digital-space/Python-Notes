# Decorators in Python

## Introduction

1️⃣ The core idea

A decorator is simply:

>A function that takes another function and returns a new function with extra behavior

So instead of modifying an existing function directly, we <span style="color: orange;">wrap it</span>.

2️⃣ Is it a nested function?

Yes — internally, decorators almost always use nested functions (closures).

Why?

Because:
- we need to define a new function inside it
- that inner function can access the original function (`func`)
- and add behavior before/after calling it

Example

```python
def add_logging(func):
    def wrapper():
        print("Starting order process...")
        func()
        print("Order process completed.")
    return wrapper

def place_order():
    print("Order placed successfully")

enhanced_function = add_logging(place_order)
enhanced_function()
```

Here:
- `add_logging` → decorator
- `wrapper` → nested function
- `func` → original function

Note that we did not do any changes to the orginal function `order_food`.

👉 What happens:
- we pass `place_order` into decorator
- decorator returns `wrapper`
- we call the new function



3️⃣ What the wrapper actually becomes

Conceptually, this happens:

```python
def enhanced_function():
    print("Starting order process...")
    print("Order placed successfully")
    print("Order process completed.")
```

4️⃣ What problem does it solve?

Decorators are commonly used when we want to:

✔️ Do something BEFORE a function
- logging
- authentication
- validation

✔️ Do something AFTER a function
- cleanup
- logging result
- sending notifications

✔️ Do BOTH

5️⃣ The shortcut syntax: `@decorator`

Instead of writing:

```python
enhanced_function = add_logging(place_order)
```

We write:

```python
@add_logging
def place_order():
    print("Order placed successfully")
```

👉 Python internally converts this to:

```python
place_order = add_logging(place_order)
```

```python
def add_logging(func): # decorator name → custom (can be any name)
    def wrapper():     # inner function → custom (can be any name)
        print("Starting order process...")
        func()         # parameter → custom (can be any name)
        print("Order process completed.")
    return wrapper

@add_logging           # @ → Python syntax for applying decorator
def place_order():
    print("Order placed successfully")

place_order()
```

Real-world use cases
- Logging
- Authentication
- Authorization
- Caching
- Timing performance
- Input validation

## Decorators with arguments

Decorators with arguments add one more level of nesting.

1️⃣ Basic example (single parameter)

Repeat order confirmation N times

```python
def repeat_order(times):
    def decorator(func):
        def wrapper(arg):
            for _ in range(times):
                func(arg)
        return wrapper
    return decorator

@repeat_order(2)
def confirm_order(order_id):
    print(f"Order {order_id} confirmed")

confirm_order("ORD101")
```

2️⃣ Why extra nesting is needed

When we write:

```python
@repeat(3)
def greet():
    ...
```

Python does:

```python
greet = repeat(3)(greet)
```

So:
- `repeat(3)` → must return a decorator
- that decorator → takes the function
- and returns the wrapper

👉 That’s why we need 3 layers:

<pre>
outer function (takes decorator args)
    ↓
decorator function (takes function)
    ↓
wrapper function (runs actual logic)
</pre>
